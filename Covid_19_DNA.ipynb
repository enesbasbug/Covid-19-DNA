#### Outline

```python
from Bio import SeqIO

# Reading our fasta file
covid_record = SeqIO.read("covid_sequence_MT385497.fasta","fasta")
covid_record
```

```python
# Seq
covid_dna = covid_record.seq
covid_dna
```

```python
# len of DNA
len(covid_dna)
```

```python
# Transcription
# DNA to mRNA
covid_mrna = covid_dna.transcribe()
covid_mrna
```

```python
# Translation
# mRNA to Amino Acid
covid_protein = covid_mrna.translate()
```

```python
# Protein / AA 
covid_protein
```

### DNA Composition
+ GC Content
+ AC Content

```python
covid_dna
```

```python
# GC Content
from Bio.SeqUtils import GC,GC123,GC_skew

GC(covid_dna)
```

```python
# AT Content
def at_content(seq):
    result = float(seq.count('A') + seq.count('T'))/len(covid_dna) * 100
    return result

at_content(covid_dna)
```

```python
# GC Skew
# Lagging or leading strand
GC_skew(covid_dna)
```

```python
from Bio.SeqUtils import MeltingTemp as mt
mt.Tm_GC(covid_dna, strict=False)
```
